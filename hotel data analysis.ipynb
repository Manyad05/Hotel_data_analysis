```python
# Rule 2: Analyze how lead time affects cancellation rates
lead_time_bins = [0, 30, 60, 90, 120, 150, float('inf')]  # Define lead time bins
lead_time_labels = ['0-30', '31-60', '61-90', '91-120', '121-150', '151+']

# Create a new column to categorize lead time into bins
ht['lead_time_bins'] = pd.cut(ht['lead_time'], bins=lead_time_bins, labels=lead_time_labels)

# Calculate cancellation rates for each lead time bin
cancellation_rates_by_lead_time = ht.groupby('lead_time_bins')['is_canceled'].mean() * 100
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate cancellation percentages
cancelled_per = ht['is_canceled'].value_counts(normalize=True)

# Set a seaborn style
sns.set(style="whitegrid")

# Create a barplot
plt.figure(figsize=(8, 6))
ax = sns.barplot(x=cancelled_per.index, y=cancelled_per.values, palette="Blues_d")

# Add labels and title
plt.title('Reservation Cancellation Status')
plt.xlabel('Status')
plt.ylabel('Percentage')

# Add percentage values on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height() * 100:.2f}%', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(cancellation_rates_by_lead_time.index, cancellation_rates_by_lead_time.values)
plt.xlabel('Lead Time (days)')
plt.ylabel('Cancellation Rate (%)')
plt.title('Cancellation Rate by Lead Time')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame 'data' with a 'customer_type' column
customer_type_counts = ['customer_type'].value_counts()

labels = customer_type_counts.index
sizes = customer_type_counts.values

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.4))
plt.title('Distribution of Customer Types')

# Draw a white circle at the center to create the donut effect
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.axis('equal')

plt.show()
```
